###ğŸ¹ğŸ¾<font color = 'blue'> **Write as the excel format after combining multiple data frames**


```python
#ì—¬ëŸ¬ê°œì˜ excel íŒŒì¼ì„ íŒŒì¼ í•˜ë‚˜ë¡œ ì €ì¥í•˜ê¸°

import pandas as pd

data1 = {'name': ['Emily', 'Andy', 'John'],
        'c++':['A', 'A+', 'B'],
        'basic': ['C', 'B', 'B+'],
        'JAVA': ['B+', 'C', 'C+']
        }

data2 = {'c0':[1,2,3],
         'c1':[4,5,6],
         'c2':[7,8,9],
         'c3':[10,11,12],
         'c4':[13,14,15]
         }

df1=pd.DataFrame(data1)
df1.set_index('name', inplace=True)
print(df1)
print('\n')

df2=pd.DataFrame(data2)
df2.set_index('c0', inplace=True)
print(df2)

#df1ì„ sheet1ìœ¼ë¡œ, df2ë¥¼ sheet2ë¡œ ì €ì¥. íŒŒì¼ëª…ì€ df_excelwriter.xlsx ë¡œ í•´ë³¸ë‹¤.

writer = pd.ExcelWriter("./df_excelwriter.xlsx")

df1.to_excel(writer, sheet_name = 'grading')
df2.to_excel(writer, sheet_name='numbers')
writer.close() #To address this, you can use the writer.close() method instead of save(). This will properly close the Excel file and ensure that all the data is saved.
```
